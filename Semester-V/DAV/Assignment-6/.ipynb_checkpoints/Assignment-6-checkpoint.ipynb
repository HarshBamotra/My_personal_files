**Q5. Consider the following DataFrame ADM containing data of freshly admitted students in a college during various rounds of admission. The DataFrame consists of the student’s name, cutoff list in which he/she has taken admission, date of admission, his/her % of marks, course code and gender.**

```python
adm = pd.read_csv("Q5_Data.csv")
adm
```

**a) Set the first column ‘Sid’ as the row index of the given DataFrame ADM. Create a pivot table of the DataFrame to display the total number of admissions as per ‘Course Code’ and  ‘Gender’.**

```python
adm.set_index("Sid", inplace = True)
adm
```

```python
adm.pivot_table(adm, index=["Course Code", "Gender"], aggfunc="count")["DateAdm"]
```

**b) For each ‘List’, find the total number of admissions, minimum ‘Marks%’and maximum ‘Marks%’ in each course.**

```python
df1 = pd.DataFrame(adm.pivot_table(adm, index=["Course Code"], aggfunc="count")["DateAdm"])
df1.columns = ["Total admissions"]
df1
```

```python
df2 = pd.DataFrame(adm.pivot_table(adm, index=["Course Code"], aggfunc="max")["Marks%"])
df2.columns = ["Maximun %"]
df2
```

```python
df3 = pd.DataFrame(adm.pivot_table(adm, index=["Course Code"], aggfunc="min")["Marks%"])
df3.columns = ["Minimun %"]
df3
```

```python
result = pd.concat([df1, df2, df3], axis = 1)
result
```

**c) Calculate and display the average ‘Marks%’ of all Female students of course ‘C112’.**

```python
print("Average marks of female students of course C112 ::", adm.pivot_table(adm, index = ["Course Code", "Gender"], aggfunc = "mean")["Marks%"].loc["C112", "Female"])
```

**c) Create a DataFrame DS with two columns ‘Dates’ and ‘Sale’ containing all dates of January 2021 and 31 random integers between 500 and 1000 respectively. Add another column ‘MovingAvg’ to DS containing the rolling average of 5 consecutive values in the ‘Sale’ column. Plot simple line plots between ‘Dates’ and ‘Sale’ as well as ‘Dates’ and ‘MovingAvg’. Explain the  utility of the rolling method  with respect to these plots.**

```python
DS = pd.DataFrame(pd.Series(np.random.randint(500,1000,size=(31)),index=pd.date_range(start='01/01/2021',periods=31)), columns = ["Sale"])
DS
```

```python
DS.index.name = "Dates"
```

```python
DS['Moving Avg']=DS.Sale.rolling(5).mean()
DS
```

```python
DS['Sale'].plot(label='Sale')
DS['Moving Avg'].plot(label='Moving Avg')
plt.legend(loc='upper left')
plt.xlabel('Dates')
plt.ylabel('Sales')
plt.title('Plot between Sales and Moving Average of Sales')
plt.show()
```

**Q7. Consider a data frame containing data about studentsi.e. name, gender and passing division:**

```python
data = pd.read_csv("Q7_Data.csv")
data = data.drop("Unnamed: 0", axis = 1)
data
```

**a) Perform one hot encoding of the last two columns ofcategorical data using the get_dummies() function.**

```python
hot_encoded = pd.get_dummies(data, columns = ["Gender", "Pass_Division"])
hot_encoded
```

**b) Sort this data frame on the “Birth Month” column (i.e. January to December). Hint: Convert Month to Categorical.**

```python
result = data.copy()
result["Birth Month"] = pd.to_datetime(data["Birth Month"].apply(lambda x: x[:3]), format = '%b')
cat = pd.date_range('2011-01-1', periods=12, freq='M').strftime('%b')
result["Birth"] = pd.Categorical(result["Birth Month"].dt.strftime('%b'), categories = cat)
result = result.sort_values(["Birth Month"])
result["Birth Month"] = result["Birth Month"].dt.month_name()
result.reset_index(drop = True)
```
