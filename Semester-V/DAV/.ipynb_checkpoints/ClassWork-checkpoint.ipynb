```python
import numpy as np
import pandas as pd
import random
```

```python
df = pd.DataFrame(np.random.randint(101, size=(50, 3)), columns = ['First', 'Second', 'Third'])
df.head()
```

```python
l = [(r , c) for r in range(df.shape[0]) for c in range(df.shape[1])]
l1 = random.sample(l, int(len(l)*(0.1)))
```

```python
for i, j in l1:
    df.iloc[i, j] = None
df.head()
```

**Number of NaN values in each columns**

```python
df.isna().sum()
```

**Drop column with NaN values greater than 5**

```python
l2 = [True if(i>5) else False for i in df.isna().sum()]
l3 = df.isna().sum().index[l2]
l3
```

```python
df = df.drop(l3, axis=1)
df
```

**Drop entry with maximum value**

```python
m = max(df.sum(axis = 1))
s = df.sum(axis = 1)
df = df.drop(s[s == m].index)
df.head()
```
