**Importing required libraries**

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

### *Q. Run Apriori algorithm to find frequent itemsets and association rules :- <br><br>1. Use minimum support as 50% and minimum confidence as 75%.*

**Reading data from csv file**

```python
df = pd.read_csv('basket_analysis.csv', index_col=0) 
df.head(5)
```

**Calculating frequent items using minimum support as 50%**

```python
freq_items = apriori(df, min_support=0.05, use_colnames=True) 
freq_items
```

**Calculating association rule using minimum confidence as 75%**

```python
association_rules(freq_items, metric="confidence", min_threshold=0.075)
```
